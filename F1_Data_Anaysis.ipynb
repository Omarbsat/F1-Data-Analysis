# **Web-Scraping**

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
r = requests.get("https://www.formula1.com/en/results.html/2021/races/1064/bahrain/race-result.html")
soup = bs(r.content)

table = soup.select("table.resultsarchive-table")[0]
columns = table.find("thead").find_all("th")
column_names = [c.string for c in columns]

table_rows = table.find("tbody").find_all("tr")
l = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [str(tr.get_text()).strip() for tr in td]
    l.append(row)
df1 = pd.DataFrame(l, columns=column_names)
df1["Date"] = "28 March"
df1["Track"] = "BAHRAIN"
df1
```

## **Data Cleaning and Wrangling**

```python
df = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12,df13,df14,df15,df16,df17,df18,df19,df20,df21,df22])
df.set_index("Pos",inplace=True)
df["First Name"] = df["Driver"].str[-3:]
del df["Driver"]
del df["Time/Retired"]
del df["No"]
df["Laps"] = pd.to_numeric(df["Laps"], errors='ignore')
df["PTS"] = pd.to_numeric(df["PTS"], errors='ignore')
clist = list(df.columns)
clist_new = clist[-1:]+clist[:-1]
df = df[clist_new]
df.rename(columns={'First Name': 'Driver'}, inplace=True)
df.head(5) 
```

# **Data** **Visualization**

```python
import matplotlib.pyplot as plt
keys = [driver for driver, df in df.groupby(['Driver'])]

plt.bar(keys,df.groupby(['Driver']).sum()['PTS'])
plt.ylabel('PTS')
plt.xlabel('Driver')
plt.xticks(keys, rotation='vertical', size=12)
plt.show()
```
